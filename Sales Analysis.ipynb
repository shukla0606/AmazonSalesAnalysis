group by , the group by function in panas is used to group data based on one or more columns in a dataframe


```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty)
```

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']  #df data me column- data ko liya 
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=20,edgecolor='Red')  # bin for  distance 
plt.xticks(rotation=90)   # for rotaiton of lables
plt.show()
```

```python
#checking b2b data by using pie chart
B2B_Check=df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')

plt.show()
```

```python
# with lable b2B buyers and retailer buyers

#checking b2b data by using pie chart
B2B_Check=df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')

plt.show()
 
```

Note ; from above chart we can se that maximum 99.2% of buyers are retialers an
d 0.8% are b2b buyerss

```python
#Sxatter plot 
x_data=df['Category']
y_data=df['Size']

#plot the scatter plot

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scater plot')
plt.show()
```

```python
# states wise data where order will be given
plt.figure(figsize=(12,10))
sns.countplot(data=df,x='ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python

# showing only top 10 state
# in the varilable ,coulum se 10 value liya 
top_10_state=df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,10))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

Note : from the above grap you can see that most of the buyers are maharastra state
